Task a

```python
import pandas as pd

data = pd.read_csv("volume_pressure_data.csv")

print(f"The first 10 rows of data are: \n {data.head(10)}")

df_volume = data['Volume']
df_pressure = data['Pressure']

mean_V = df_volume.mean()
mean_P = df_pressure.mean()

print(f"The mean volume is {mean_V}, while the mean pressure is {mean_P}")

```

Task b

```python
import matplotlib.pyplot as plt

volume = data['Volume']
pressure = data['Pressure']

plt.plot(volume, pressure)
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.title('Volume vs Pressure')
plt.grid(True)
plt.show()
```
